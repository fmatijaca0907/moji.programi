Ova aplikacija treba uzeti datum kad ste ulili gorivo
1.Pita kosrinika da li je ovaj mjesec utocio gorivo, ako je dodaje datum, ako nije


```python
import datetime as dt

def take_date_today() ->dt.date:
    date = dt.date.today()
    return date

def today_y() -> dt.date.year :
    date = take_date_today()
    year = date.year
    return year

def today_m():
    date = take_date_today()
    month = date.month
    return month

def today_d():
    date = take_date_today()
    day = date.day
    return day

def answer_gas_m():
    what_month = input("Did you fill up with gas this month?  (yes/no)")
    what_month = what_month.lower()
    potvrda = "yes"
    potvrda1 = "no"
    while True:
        if what_month == potvrda:
            break
        elif what_month == potvrda1:
            break
        else:
            what_month = input("You have to enter yes or no.")
            what_month = what_month.lower()

    if what_month == potvrda:
        what_month = True
    else:
        what_month = False

    return what_month

def fill_month():
    fill_gas = answer_gas_m()
    if fill_gas:
        month = today_m()
        
    else:
        month = input("Unesi mjesec: (npr.08)")
        confirmation = month.isdigit()
        while confirmation != True:
            month = input("Unesi broj! ")
            confirmation = month.isdigit()
        month = int(month)

    return month
    
        
def answer_gas_d():
    what_day = input("Did you fill up with gas today?  (yes/no)")
    what_day = what_day.lower()
    potvrda = "yes"
    potvrda1 = "no"
    while True:
        if what_day == potvrda:
            break
        elif what_day == potvrda1:
            break
        else:
            what_day = input("You have to enter yes or no.")
            what_day = what_day.lower()

    if what_day == potvrda:
        what_day = True
    else:
        what_day = False

    return what_day

def fill_day():
    fill_gas = answer_gas_d()
    if fill_gas:
        day = today_d()
        
    else:
        day = input("Unesi dan: (npr.24)")
        confirmation = day.isdigit()
        while confirmation != True:
            day = input("Unesi broj! ")
            confirmation = day.isdigit()
        day = int(day)

    return day

def car_fill_date():
    year = today_y()
    month = fill_month()
    day = fill_day()
    return year,month,day

car_fill_date()
```
